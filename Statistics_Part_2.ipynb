```python
#Answer 1

import numpy as np
import scipy.stats as stats

def z_test(sample_mean, pop_mean, pop_std, sample_size, alpha=0.05):
    # Compute the Z-score
    standard_error = pop_std / np.sqrt(sample_size)
    z_score = (sample_mean - pop_mean) / standard_error

    # Compute the p-value for two-tailed test
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    # Print results
    print(f"Z-Score: {z_score:.4f}")
    print(f"P-Value: {p_value:.4f}")

    # Decision
    if p_value < alpha:
        print("Reject H0: The sample mean is significantly different from the population mean.")
    else:
        print("Fail to Reject H0: No significant difference between sample and population mean.")

# Given data
sample_mean = 52500  # Sample mean salary
pop_mean = 50000     # Population mean salary
pop_std = 8000       # Population standard deviation
sample_size = 30     # Sample size

# Perform Z-test
z_test(sample_mean, pop_mean, pop_std, sample_size)

```

```python
#Answer 5

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def visualize_type1_type2(pop_mean=100, sample_mean=105, std_dev=15, sample_size=30, alpha=0.05):
    """
    Visualizes Type 1 and Type 2 errors in hypothesis testing.

    Parameters:
    - pop_mean: Population mean (H0)
    - sample_mean: Sample mean (HA)
    - std_dev: Population standard deviation
    - sample_size: Size of the sample
    - alpha: Significance level
    """
    # Compute standard error
    standard_error = std_dev / np.sqrt(sample_size)

    # Compute critical value for alpha
    critical_value = stats.norm.ppf(1 - alpha)  # One-tailed test

    # Compute beta (Type 2 error) for alternative hypothesis
    beta = stats.norm.cdf(critical_value, loc=sample_mean, scale=standard_error)

    # X-axis range for visualization
    x = np.linspace(pop_mean - 4 * std_dev, pop_mean + 4 * std_dev, 1000)

    # Normal distributions for H0 and HA
    y_h0 = stats.norm.pdf(x, pop_mean, standard_error)  # Null hypothesis
    y_ha = stats.norm.pdf(x, sample_mean, standard_error)  # Alternative hypothesis

    plt.figure(figsize=(10, 6))

    # Plot H0 distribution
    plt.plot(x, y_h0, label="H0: Null Distribution", color='blue')
    plt.fill_between(x, y_h0, where=(x > critical_value * standard_error + pop_mean), color='red', alpha=0.5, label="Type 1 Error (Œ±)")

    # Plot HA distribution
    plt.plot(x, y_ha, label="HA: Alternative Distribution", color='green')
    plt.fill_between(x, y_ha, where=(x < critical_value * standard_error + pop_mean), color='purple', alpha=0.5, label="Type 2 Error (Œ≤)")

    # Vertical lines for critical value
    plt.axvline(x=critical_value * standard_error + pop_mean, color='black', linestyle='dashed', label=f"Critical Value = {critical_value * standard_error + pop_mean:.2f}")

    # Labels and legend
    plt.title("Type 1 and Type 2 Errors in Hypothesis Testing")
    plt.xlabel("Test Statistic")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.show()

    print(f"Critical Value (ZŒ±): {critical_value:.2f}")
    print(f"Type 2 Error (Œ≤): {beta:.4f}")
    print(f"Power of the test (1 - Œ≤): {1 - beta:.4f}")

# Call function to visualize Type 1 & Type 2 errors
visualize_type1_type2()
```

```python
#Answer 8

import numpy as np
import scipy.stats as stats

def perform_z_t_tests(sample, pop_mean, pop_std, alpha=0.05):
    """
    Performs both a one-sample Z-test and a one-sample T-test on the same dataset.

    Parameters:
    - sample: Sample dataset (numpy array)
    - pop_mean: Known population mean
    - pop_std: Known population standard deviation (used for Z-test)
    - alpha: Significance level (default = 0.05)
    """

    # Sample statistics
    sample_mean = np.mean(sample)
    sample_size = len(sample)
    sample_std = np.std(sample, ddof=1)  # Sample standard deviation (unbiased)

    # Compute standard errors
    standard_error_z = pop_std / np.sqrt(sample_size)  # For Z-test
    standard_error_t = sample_std / np.sqrt(sample_size)  # For T-test

    # Perform Z-test
    z_score = (sample_mean - pop_mean) / standard_error_z
    p_value_z = 2 * (1 - stats.norm.cdf(abs(z_score)))  # Two-tailed test

    # Perform T-test
    t_stat, p_value_t = stats.ttest_1samp(sample, pop_mean)

    # Print Results
    print(f"Sample Size: {sample_size}")
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Population Mean: {pop_mean:.2f}")
    print(f"Population Std Dev: {pop_std:.2f}")

    print("\n--- Z-Test Results ---")
    print(f"Z-Score: {z_score:.4f}")
    print(f"P-Value (Z-test): {p_value_z:.4f}")
    if p_value_z < alpha:
        print("Reject H0: Significant difference found (Z-test).")
    else:
        print("Fail to Reject H0: No significant difference (Z-test).")

    print("\n--- T-Test Results ---")
    print(f"T-Statistic: {t_stat:.4f}")
    print(f"P-Value (T-test): {p_value_t:.4f}")
    if p_value_t < alpha:
        print("Reject H0: Significant difference found (T-test).")
    else:
        print("Fail to Reject H0: No significant difference (T-test).")

# Simulate a sample dataset (e.g., test scores)
np.random.seed(42)  # For reproducibility
sample_data = np.random.normal(loc=75, scale=10, size=25)  # Sample size < 30 (small)

# Known population parameters
population_mean = 78
population_std = 10  # Known for Z-test

# Perform both Z-test and T-test
perform_z_t_tests(sample_data, population_mean, population_std)
```

```python
#Answer 9

import numpy as np
import scipy.stats as stats

def confidence_interval(sample, confidence=0.95):
    """
    Calculates the confidence interval for a sample mean.

    Parameters:
    - sample: List or numpy array of sample data
    - confidence: Confidence level (default = 95%)

    Returns:
    - Tuple containing lower and upper bounds of the confidence interval
    """
    n = len(sample)  # Sample size
    mean = np.mean(sample)  # Sample mean
    std_dev = np.std(sample, ddof=1)  # Sample standard deviation (unbiased)

    # Get the critical t-value for the given confidence level
    t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)

    # Compute margin of error
    margin_of_error = t_critical * (std_dev / np.sqrt(n))

    # Confidence interval bounds
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return lower_bound, upper_bound

# Simulate sample data (e.g., student test scores)
np.random.seed(42)
sample_data = np.random.normal(loc=75, scale=10, size=30)  # Mean=75, Std=10, Sample=30

# Compute confidence interval
ci_lower, ci_upper = confidence_interval(sample_data)

# Print results
print(f"Sample Mean: {np.mean(sample_data):.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
#Answer 10

import numpy as np
import scipy.stats as stats

def margin_of_error(sample, confidence=0.95):
    """
    Calculates the margin of error for a given sample and confidence level.

    Parameters:
    - sample: List or numpy array of sample data
    - confidence: Confidence level (default = 95%)

    Returns:
    - Margin of Error (MoE)
    """
    n = len(sample)  # Sample size
    std_dev = np.std(sample, ddof=1)  # Sample standard deviation (unbiased)

    # Get the critical t-value for the given confidence level
    t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)

    # Compute margin of error
    moe = t_critical * (std_dev / np.sqrt(n))

    return moe

# Simulate sample data (e.g., survey responses)
np.random.seed(42)
sample_data = np.random.normal(loc=75, scale=10, size=30)  # Mean=75, Std=10, Sample=30

# Compute margin of error
moe_value = margin_of_error(sample_data)

# Print results
print(f"Sample Size: {len(sample_data)}")
print(f"Sample Standard Deviation: {np.std(sample_data, ddof=1):.2f}")
print(f"95% Margin of Error: ¬±{moe_value:.2f}")
```

```python
#Answer 11

def bayes_theorem(prior, likelihood, false_positive):
    """
    Computes the posterior probability using Bayes' Theorem.

    Parameters:
    - prior: P(H) - Prior probability of hypothesis (having the disease)
    - likelihood: P(E|H) - Probability of evidence given hypothesis (test accuracy)
    - false_positive: P(E|¬¨H) - Probability of false positive (test error rate)

    Returns:
    - Posterior probability P(H|E)
    """
    # Compute total probability of evidence P(E)
    evidence = (likelihood * prior) + (false_positive * (1 - prior))

    # Compute posterior probability using Bayes' theorem
    posterior = (likelihood * prior) / evidence

    return posterior

# Given values
prior_prob = 0.01  # 1% of people have the disease
likelihood_prob = 0.90  # 90% accuracy for those with the disease
false_positive_prob = 0.05  # 5% false positives for healthy people

# Calculate posterior probability
posterior_prob = bayes_theorem(prior_prob, likelihood_prob, false_positive_prob)

# Display result
print(f"Probability of having the disease given a positive test: {posterior_prob:.4f} ({posterior_prob*100:.2f}%)")
```

```python
#Answer 16

import numpy as np
import scipy.stats as stats

def perform_f_test(sample1, sample2, alpha=0.05):
    """
    Perform an F-test to compare the variances of two independent samples.

    Parameters:
    - sample1: First dataset (numpy array or list)
    - sample2: Second dataset (numpy array or list)
    - alpha: Significance level (default: 0.05)

    Returns:
    - F-statistic and P-value
    """
    var1 = np.var(sample1, ddof=1)  # Sample variance (unbiased)
    var2 = np.var(sample2, ddof=1)

    # Ensure F-statistic is always >1 by setting larger variance in numerator
    if var1 > var2:
        F_stat = var1 / var2
        dfn, dfd = len(sample1) - 1, len(sample2) - 1  # Degrees of freedom
    else:
        F_stat = var2 / var1
        dfn, dfd = len(sample2) - 1, len(sample1) - 1

    # Compute p-value (two-tailed test)
    p_value = 2 * (1 - stats.f.cdf(F_stat, dfn, dfd))

    # Display results
    print(f"F-statistic: {F_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Interpretation
    if p_value < alpha:
        print("\nReject the null hypothesis: Variances are significantly different.")
    else:
        print("\nFail to reject the null hypothesis: No significant difference in variances.")

# Simulate two random samples
np.random.seed(42)  # For reproducibility
sample1 = np.random.normal(loc=50, scale=10, size=30)  # Mean=50, Std Dev=10, Size=30
sample2 = np.random.normal(loc=50, scale=15, size=30)  # Mean=50, Std Dev=15, Size=30

# Perform F-test
perform_f_test(sample1, sample2)
```

```python
#Answer 19

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.stats.diagnostic import het_breuschpagan
from statsmodels.formula.api import ols
from statsmodels.stats.api import anova_lm

def check_anova_assumptions(data, groups):
    """
    Check the assumptions for ANOVA: Normality, Independence, and Equal Variance (Homogeneity of variance).

    :param data: List or array-like of numerical values representing the response variable
    :param groups: List or array-like of group labels corresponding to the data
    :return: Prints results and plots for assumption checks
    """

    # 1. Check for Normality within each group (using Shapiro-Wilk Test)
    print("Checking Normality (Shapiro-Wilk Test):")
    unique_groups = np.unique(groups)
    for group in unique_groups:
        group_data = data[np.array(groups) == group]
        stat, p_value = stats.shapiro(group_data)
        print(f"Group {group}: Shapiro-Wilk p-value = {p_value}")
        if p_value < 0.05:
            print(f"Warning: Group {group} may not follow a normal distribution.")

    # 2. Check for Independence (visually with scatter plot or residuals plot)
    # Independence assumption usually comes from study design, so let's assume this
    # is validated externally. But we can check residuals plot if model is already fitted.
    print("\nChecking for Independence: This is typically validated by study design.")

    # 3. Check for Equal Variance (Homogeneity of variance) using Levene's Test
    print("\nChecking Equal Variance (Levene's Test):")
    stat, p_value = stats.levene(*[data[np.array(groups) == group] for group in unique_groups])
    print(f"Levene's Test p-value = {p_value}")
    if p_value < 0.05:
        print("Warning: Variance across groups may not be equal.")

    # Optional: Plot histograms to visually check normality assumption
    print("\nVisualizing distributions for normality check:")
    sns.boxplot(x=groups, y=data)
    plt.show()

# Example usage:
# Assuming you have data as a list/array and groups as a list/array of the same length
data = np.random.randn(100)  # Example data (normal distribution)
groups = ['Group1'] * 50 + ['Group2'] * 50  # Example group labels

check_anova_assumptions(data, groups)

```

```python
#Answer 25

import numpy as np
import scipy.stats as stats

def z_test_proportions(x1, n1, x2, n2):
    # Sample proportions
    p1 = x1 / n1
    p2 = x2 / n2

    # Pooled proportion
    p_pooled = (x1 + x2) / (n1 + n2)

    # Standard error
    se = np.sqrt(p_pooled * (1 - p_pooled) * (1/n1 + 1/n2))

    # Z-score
    z_score = (p1 - p2) / se

    # P-value (two-tailed test)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    return z_score, p_value

# Example Data: Group 1 (successes: 40, total: 200), Group 2 (successes: 55, total: 220)
x1, n1 = 40, 200  # Group 1
x2, n2 = 55, 220  # Group 2

# Perform Z-test
z_stat, p_val = z_test_proportions(x1, n1, x2, n2)

# Print Results
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_val:.4f}")

# Interpretation
alpha = 0.05
if p_val < alpha:
    print("Reject H‚ÇÄ: The proportions are significantly different.")
else:
    print("Fail to reject H‚ÇÄ: No significant difference in proportions.")
```

```python
#Answer 26

import numpy as np
import scipy.stats as stats

def z_test_proportions(x1, n1, x2, n2):
    """
    Performs a Z-test for proportions to compare two independent groups.

    Parameters:
        x1 (int): Number of successes in group 1
        n1 (int): Total observations in group 1
        x2 (int): Number of successes in group 2
        n2 (int): Total observations in group 2

    Returns:
        tuple: (Z-statistic, P-value)
    """
    # Sample proportions
    p1 = x1 / n1
    p2 = x2 / n2

    # Pooled proportion
    p_pooled = (x1 + x2) / (n1 + n2)

    # Standard error
    se = np.sqrt(p_pooled * (1 - p_pooled) * (1/n1 + 1/n2))

    # Compute Z-score
    z_score = (p1 - p2) / se

    # Compute p-value (two-tailed test)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    return z_score, p_value

# Example: Compare two groups
x1, n1 = 50, 200  # Group 1: 50 successes out of 200
x2, n2 = 60, 220  # Group 2: 60 successes out of 220

# Perform Z-test
z_stat, p_val = z_test_proportions(x1, n1, x2, n2)

# Print results
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_val:.4f}")

# Interpretation
alpha = 0.05
if p_val < alpha:
    print("Reject H‚ÇÄ: The proportions are significantly different.")
else:
    print("Fail to reject H‚ÇÄ: No significant difference in proportions.")
```
